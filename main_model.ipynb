```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_data = pd.read_csv("./playground-series-s4e10/train.csv")
test_data = pd.read_csv("./playground-series-s4e10/test.csv")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, classification_report
```

Making categorical Columns into numericals

```python
def preprocess_data(df):
    le = LabelEncoder()
    categorical_columns = ['person_home_ownership', 'loan_intent', 'loan_grade', 'cb_person_default_on_file']

    for col in categorical_columns:
        df[col] = le.fit_transform(df[col])
    return df
```

```python
train_data_transformed = preprocess_data(train_data)
test_data_transformed = preprocess_data(test_data) 
```

```python
X_train = train_data_transformed.drop(['id', 'loan_status'], axis=1)
Y_train = train_data_transformed['loan_status']
```

```python
X_Test = test_data_transformed.drop(['id'], axis=1)
```

```python
X_Test
```

```python
X_train
```

```python
model = RandomForestClassifier(n_estimators=300, random_state=42)
model.fit(X_train, Y_train)
```

```python
Y_predictions = model.predict(X_Test)
```

```python
result_df = pd.DataFrame(
    {
        "id" : test_data_transformed['id'],
        "loan_status" : Y_predictions
    }
)
```

```python
result_df.to_csv("submission2.csv", index=False)
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
import numpy as np

# Define the parameter grid
param_grid = {
    'n_estimators': [100, 200, 300, 400, 500],
    'max_depth': [10, 20, 30, 40, 50, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2']  # Remove 'auto'
}

# Create a base model
rf = RandomForestClassifier(random_state=42)

# Instantiate the random search model
random_search = RandomizedSearchCV(estimator=rf, param_distributions=param_grid, 
                                   n_iter=100, cv=5, verbose=2, random_state=42, 
                                   n_jobs=-1, error_score='raise')

# Fit the random search model
random_search.fit(X_train, Y_train)

# Get the best parameters
print(random_search.best_params_)

# Use the best model found
best_rf = random_search.best_estimator_
```

```python
y_predictions = best_rf.predict(X_Test)
```

```python
result_df_2 = pd.DataFrame(
    {
        "id" : test_data_transformed['id'],
        "loan_status" : y_predictions
    }
)
```
